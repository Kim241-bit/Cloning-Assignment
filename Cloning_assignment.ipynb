```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection 

# Load Titanic dataset
df = sns.load_dataset('titanic')
print(df.head())
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count values for a categorical variable (e.g., class)
class_counts = df['class'].value_counts()

# Plot pie chart
plt.figure(figsize=(7, 7))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Passenger Classes in Titanic')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['fare'].dropna(), bins=30, kde=True, color="blue")
plt.title("Titanic Passenger Fare Distribution")
plt.xlabel("fare")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="sex", y="survived", data=df, estimator=lambda x: sum(x) / len(x))
plt.title("Survival Rate by Sex")
plt.xlabel("Passenger Sex")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)  # Since it's a probability (0 to 1)
plt.show()
```
